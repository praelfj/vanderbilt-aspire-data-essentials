# Hospice data set
## import hospice data

```python
cancer_deaths = pd.read_csv('./data/Hospice_General_Information.csv')
```

## find TN

```python
tn_cancer_deaths = cancer_deaths[cancer_deaths['State'] == 'TN']
tn_cancer_deaths.head()
```

## subset columns and make column case smaller

```python
tn_cancer_deaths = tn_cancer_deaths[['Facility Name', 'Address Line 1', 'City', 'State', 
                                     'County Name', 'CMS Region', 'Ownership Type', 'Certification Date']]
tn_cancer_deaths.head()
```

```python
tn_cancer_deaths.columns = tn_cancer_deaths.columns.str.lower()
tn_cancer_deaths.head()
```

## examine head and tail of TN data; print dimensions of data

```python
tn_cancer_deaths.head()
```

```python
tn_cancer_deaths.tail()
```

```python
print('there are ' + str(tn_cancer_deaths.shape[0]) + ' hospice facilities in TN')
```
